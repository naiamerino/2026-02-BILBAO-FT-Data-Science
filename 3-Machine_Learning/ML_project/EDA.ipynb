```python
# los nombres vienen con patrocinadores. clasificarlos con las palabras clave
# si alguna (any) de estas palabras (x) está contenida (in) en el nombre del torneo (n)

def clasificar_torneo(nombre):
    n = nombre.lower()
    
    # Grand Slams — nombres estables
    if any(x in n for x in ['australian open', 'roland garros', 'french open', 
                              'wimbledon', 'us open']):
        return 'GS'
    
    # WTA Finals — varios nombres históricos
    if any(x in n for x in ['wta finals', 'wta elite trophy', 'championships',
                              'tournament of champions', 'riyadh finals', 'wta tour championships']):
        return 'WTA_Finals'
  
      # WTA 1000 — los torneos grandes de siempre, por ciudad
    if any(x in n for x in ['indian wells', 'miami', 'madrid', 'roma', 'rome',
                              'internazionali', 'canada', 'toronto', 'rogers',
                              'canadian', 'cincinnati', 'beijing', 'china open',
                              'wuhan', 'doha', 'qatar', 'dubai', 'bnp paribas open',
                              'national bank open', 'sony ericsson open', 
                              'western & southern financial group','shenzhen']):
        return 'WTA1000'
    
    # WTA 500 
    if any(x in n for x in ['abu dhabi','abu dabi', 'adelaide', 'berlin', 'bett1open', 'eastbourne',
                              'Rothesay', 'bad homburg', 'san jose', 'silicon valley',
                              'guangzhou', 'osaka', 'tokyo','toray pan pacific','seoul',
                              'strasbourg', 'birmingham', 'nottingham', 'brisbane',
                              'linz', 'merida', 'monterrey', 'seoul', 'charleston',
                              'stuttgart', 'porsche','washington','mubadala', 'guadalajara',
                              'gdl open akron','family circle cup', 'aegon', 'kremlin',
                              'pilot pen', 'new haven', 'bank of the west',
                              'eastbourne', 'rothesay','birmingham', 'sydney international', 'stanford',
                              'shenzhen', 'luxembourg', 'belgium', 'diamond games', 'fortis', 'san diego',
                              'zhengzhou', 'german']):
        return 'WTA500'
    
    return 'WTA250_o_menor'

wta['tournament_type'] = wta['Tournament'].map(clasificar_torneo)
```

```python
def forma_reciente(df, jugadora, fecha_limite, meses=2):

    """Calcula win rate de una jugadora en los dos meses anteriores"""
    
    fecha_inicio = fecha_limite - pd.DateOffset(months=meses)
    
    # partidos donde jugó (como Player_1 o Player_2) en el tiempo indicado (2 meses antes)
    mask = (
        ((df['Player_1'] == jugadora) | (df['Player_2'] == jugadora)) &
        (df['Date'] < fecha_limite) &   
        (df['Date'] >= fecha_inicio)
    )
     
    partidos = df[mask]
    if len(partidos) == 0:
        return 0  # viene de lesión o de mucho tiempo sin competir
    
    victorias = (partidos['Winner'] == jugadora).sum()
    return victorias/len(partidos)
```

```python
def winrate (df, jugadora, fecha_limite, superficie=None, ronda= None):

    """Calcula win rate de una jugadora en una superficie o en una ronda"""
    
    # partidos donde ha jugado (como Player_1 o Player_2) históricamente
    mask = (
        ((df['Player_1'] == jugadora) | (df['Player_2'] == jugadora)) &
        (df['Date'] < fecha_limite) 
    )
    if superficie:
        mask &= (df['Surface'] == superficie)
    
    if ronda:
        mask &= (df['Round'] == ronda)

    partidos = df[mask]
    if len(partidos) == 0:
        # Tratamiento de nulos. Se entiende que no ha jugado nunca en esa superficie o esa ronda
        # # Valor "novata en esta condición" es algo peor que el 0.5 que sería lo neutro 
        return 0.4  
    
    # Cuantos ha ganado
    victorias = (partidos['Winner'] == jugadora).sum()
    return victorias/len(partidos)
    
```

```python
def headtohead (df, p1, p2, fecha_limite):

    """Calcula win rate en el head to head entre las dos jugadoras.
        Devuelve el porcentaje de victorias de la jugadora 1"""

# partidos donde se han enfrentado ANTES de la fecha límite
    mask = (
        (
            ((df['Player_1'] == p1) & (df['Player_2'] == p2)) | 
            ((df['Player_1'] == p2) & (df['Player_2'] == p1))
        ) & (df['Date'] < fecha_limite)
    )
     
    partidos = df[mask]
    if len(partidos) == 0:
        # si no se han enfrentado nunca relleno con lo neutro
        return 0.5  
    
# Aquí contamos cuántas veces ganó P1 específicamente
    victorias_p1 = (partidos['Winner'] == p1).sum()
    return victorias_p1/len(partidos)
    

```

```python
def experiencia (df, jugadora, fecha_limite):

    """Calcula el número total de partidos jugados por la jugadora"""

    mask = (
        ((df['Player_1'] == jugadora) | (df['Player_2'] == jugadora)) &
        (df['Date'] < fecha_limite) 
    )

    partidos = df[mask]
    return len(partidos)
```

```python
# Voy a crear un nuevo dataframe con las features que será el que entre al modelo

wta = wta.reset_index()

features = []
for i, partido in wta.iterrows():

    
    p1 = partido['Player_1']  
    p2 = partido['Player_2']  
    fecha = partido['Date']
    row = {
        
        # Id para luego poder hacer tracking por si acaso
        'match_id': i,
        'date': partido['Date'],
        'odds_1': partido['prob_1'],
        'odds_2': partido['prob_1'],

        #Features del partido
        'surface': partido['Surface'],
        'round': partido['Round'],

        #Features calculadas para las jugadoras
        'rank_diff': float(partido['Rank_1']) - float(partido['Rank_2']),
        'wins2meses_p1': forma_reciente(wta, p1, fecha),
        'wins2meses_p2': forma_reciente(wta, p2, fecha),
        'ratio_superficie_p1': winrate (wta, p1, fecha, superficie = partido['Surface']),
        'ratio_superficie_p2':winrate (wta, p2, fecha, superficie = partido['Surface']),
        'h2h': headtohead (wta, p1, p2, fecha),
        'ratio_ronda_p1': winrate (wta, p1, fecha, ronda = partido['Round']),
        'ratio_ronda_p2': winrate (wta, p2, fecha, ronda = partido['Round']),
        'experiencia_p1': experiencia (wta, p1, fecha),
        'experiencia_p2': experiencia (wta, p2, fecha),
        #'ratio_total_p1': winrate (wta, p1, fecha, meses=) si lo meto cambiar en app.py
        
        # Target
        'target': partido['target']  # 1 o 2
        }
    features.append(row)
    
#añadir feature de inexperiencia para dar más info a los ceros en forma. ¿es una jugadora novata?

historico_partidos = pd.DataFrame(features)
historico_partidos['is_new_p1'] = (historico_partidos['experiencia_p1'] < 10).astype(int)
historico_partidos['is_new_p2'] = (historico_partidos['experiencia_p2'] < 10).astype(int)
# Guardo el dataframe como archivo para poder recuperarlo sin tener que ejecutarlo todo
historico_partidos.to_csv ('historico_partidos.csv', index=False) 
# Para leerlo pd.read_csv ('historico_partidos.csv', parse_dates=['date']), si no pierde el formato de fecha
```
