## Creamos el modelo

```python
X = np.array(dataframe.drop(columns=['clase']))
y = np.array(dataframe['clase'])
```

```python
X = dataframe.drop("clase",axis=1).values
y = dataframe["clase"].values
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# a veces los modelos logísticos no convergen. Puede ser que no lleguen al mínimo y sigan y sigan. Por eso marcamos el número
# máximo de intentos con max_iter
model = LogisticRegression(max_iter=1000)
model.fit(X, y)
```

```python
# No hemos hecho train-test porque para enseñar. Pero habría que hacerlo
predictions = model.predict(X)
print(predictions)
```

```python
predicions_proba = model.predict_proba(X)
print(np.round(np.array(predicions_proba), 2))
#predicions_proba -- > devuelve por cada entrada la probabilidad de que sea de cada una de las tres clases
```

```python
#Es F1 score. métrica combinada entre precision entre 0 y 1
model.score(X, y) 
```

```python
model.classes_
```

# Validación del Modelo

```python
validation_size = 0.20
seed = 7
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,
                                                                    y,
                                                                    test_size=validation_size,
                                                                    random_state=seed)
```

```python
#CROSS VALIDATION - n_splits/k_fold
# Los datos se dividen en 10 bloques iguales (folds), y en cada vuelta uno distinto hace de test
#  mientras los 9 restantes entrenan el modelo. La división se hace una sola vez al principio,
#  y luego simplemente se va rotando cuál actúa como test.
name='Logistic Regression'
kfold = model_selection.KFold(n_splits=10) #Parte los datos en 10 trozos para usar validación cruzada / cross validation
cv_results = model_selection.cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')

msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
print(cv_results)
print(msg)
```

```python
#En los resultados de arriba se ve que según le caigan los datos hay mucha diferencia entre el F1
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, Y_train)
```

```python
predictions = model.predict(X_test)
print(accuracy_score(Y_test, predictions))
predictions
```

```python
acierto = accuracy_score(Y_test, predictions)

error = 1 - acierto
print("Acierto:", round(acierto*100, 2), "%")
print("Error:", round(error*100, 2), "%")
```
