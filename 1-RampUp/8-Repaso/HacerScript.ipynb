Para hacer un script:
Crear un archivo .py y pegar ahí el código.
Para ejecutarlo hay que ir a la consola, colocarse en el directorio y ejecutar:
python nombrearchivo.py

En caso de que haya varios módulos:
Hay que crear un archivo __init__.py
un main.py donde se ejecuta
En el principal se pone al principio qué quieres importar
from la_carpeta import loquequieras
Si está separado por modulos a la hora de ejecutarlo hay que poner el comando -m
python -m 

Para limpiar la terminal:
os.clear
