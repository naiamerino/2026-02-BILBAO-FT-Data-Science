```python
print ("¿Qué operación quieres realizar?")
print ("Elige entre las siguientes opciones:")
print ("1 - Visualizar todos los libros de la biblioteca")
print ("2 - Buscar libro" )
print ("3 - Añadir libro")
print ("4 - Eliminar libro por título")
opcion = int(input("Introduce la opción deseada:"))
match opcion:
    case 1:
        for elemento in libros:
            print (*elemento.items())
    case 2:
        titulo = input ("Introduce el titulo del libro a buscar: ")
        for elemento in libros:
            if elemento ["Titulo"] == titulo:
                print (*elemento.items())
    case 3:
        titulo = input ("Introduce el titulo del libro a añadir: ")
        autor = input ("Introduce el autor del libro a añadir")
        libros.append ({"Título": titulo, "Autor": autor, "Alquilado":False})
        for elemento in libros:
            print (*elemento.items())
    case 4:
        titulo = input ("Introduce el titulo del libro a eliminar: ")
        for elemento in libros:
            if elemento ["Titulo"] == titulo:
                elemento.clear()
        # para comprobar
        for elemento in libros:
            print (*elemento.items())
    case 5:
        titulo = input ("Introduce el titulo del libro a alquilar: ")
        for elemento in libros:
            if elemento ["Titulo"] == titulo:
                elemento ["Alquilado"] = True
        # para comprobar
        for elemento in libros:
            print (*elemento.items())
    case 6:
        titulo = input ("Introduce el titulo del libro a devolver: ")
        for elemento in libros:
            if elemento ["Titulo"] == titulo:
                elemento ["Alquilado"] = False
        # para comprobar
        for elemento in libros:
            print (*elemento.items())
            
    

            

```
