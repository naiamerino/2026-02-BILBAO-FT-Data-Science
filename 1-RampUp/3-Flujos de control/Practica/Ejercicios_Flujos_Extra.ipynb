## Ejercicio Extra
Programa una calculadora que

* Pregunte qué operación quieres realizar (multiplicación/división/suma/resta/potencia/salir)
* Pida dos números 
* Realice la operación
* Siga funcionando hasta que el usuario elija salir

#### Bonus:
* Que guarde todas las operaciones realizadas y las imprima por pantalla cuando el usuario elija esa opción
* Que sea robusta frente a entradas erróneas del usuario o frente a operaciones como dividir por 0

```python
import math
import pprint
import time
```

```python
# Mostramos el menú inicial con instrucciones
print("Hola, ¿qué operación quieres realizar?")
time.sleep(0.1)
print("Opciones disponibles:")
time.sleep(0.1)
print()
time.sleep(0.1)
print("1 - Multiplicación ✖")
time.sleep(0.1)
print("2 - División ➗")
time.sleep(0.1)
print("3 - Suma ➕")
time.sleep(0.1)
print("4 - Resta ➖")
time.sleep(0.1)
print("5 - Potencia 📈")
time.sleep(0.1)
print()
time.sleep(0.1)
print("Escribe 'salir' 🚪 para terminar.")
time.sleep(0.1)
print("Escribe la palabra 'resultados' para poder verlos 👀")
time.sleep(0.1)
print()
time.sleep(0.1)


# Inicializamos la variable usuario con una cadena vacía
usuario = ""

# Lista vacía para almacenar los resultados de las operaciones
resultados= []

# Bucle que se ejecuta mientras el usuario no escriba "salir"
while usuario != "salir":
    # Pedimos la operación al usuario
    usuario = input("\nElige una opción (1/2/3/4/5), escribe 'salir' 🚪 o escribe 'resultados' si quieres ver el total de todas las operaciones: ")

    # Si el usuario quiere salir, terminamos el bucle
    if usuario.lower() == "salir":
        print("Gracias por usar la calculadora. ¡Hasta luego! 👋")
        break  # Sale del bucle
    
    # hacemos lo mismo con resultados, cuando el usuarioescribe esa palabra, se para el bucle 
    if usuario.lower() == "resultados":
        print ("Aquí tienes los resultados de tus operaciones 👀:")
        #con el pprint nos formatea el resultado para visualizarlo mejor
        pprint.pprint(resultados)
        break #salimos del bucle para imprimir los resultados
    
    
    # Validamos que la opción ingresada sea un número válido
    if usuario not in ["1", "2", "3", "4", "5"]:
        print("❌ Opción no válida. Escribe un número entre 1 y 5 o 'salir'.")
        continue  # Volver a pedir la opción sin ejecutar más código en esta iteración

    # Convertimos la opción a entero
    usuario = int(usuario)
    
    #tengo que volver a validar los números
    entrada_1= input("Ingresa el primer número:")
    #para ello he probado el método isdigit
    #Pedimos el primer número y validamos que sea un número positivo
    if not entrada_1.isdigit():
        print("❌ Número no válido. Escribe un número positivo.")
        continue
    num_1 = int(entrada_1)
    
    # Pedimos el segundo número y validamos que sea un número positivo
    entrada_2= input("Ingresa el segundo número:")
    if not entrada_2.isdigit():
        print("❌ Número no válido. Escribe un número positivo.")
        continue
    num_2 = int(entrada_2)
 

    # Realizamos la operación correspondiente según la opción escogida
    if usuario == 1:
        resultado = num_1 * num_2
        print(f"El resultado de {num_1} ✖ {num_2} es {resultado}")
        #con este método (append) permite que más tarde, cuando el usuario elija ver los resultados, 
        #la lista resultados se imprima con todas las operaciones realizadas
        resultados.append(f"multiplicación {num_1} x {num_2} = {resultado}")

    #En la división comprobamos que el usuario no divida entre 0
    elif usuario == 2:
        if num_2 == 0:
            print("❌ Error: No se puede dividir entre 0.")
        else:
            resultado = num_1 / num_2
            print(f"El resultado de {num_1} ➗ {num_2} es {resultado}")
            resultados.append(f"división {num_1} ➗ {num_2} = {resultado}")
            
    elif usuario == 3:
        resultado = num_1 + num_2
        print(f"El resultado de {num_1} ➕ {num_2} es {resultado}")
        resultados.append(f"suma {num_1} ➕ {num_2} = {resultado}")

    elif usuario == 4:
        resultado = num_1 - num_2
        print(f"El resultado de {num_1} ➖ {num_2} es {resultado}")
        resultados.append(f"suma {num_1} ➖ {num_2} = {resultado}")


    elif usuario == 5:
        resultado = num_1 ** num_2
        print(f"El resultado de {num_1} 📈 {num_2} es {resultado}")
        resultados.append(f"suma {num_1} 📈 {num_2} = {resultado}")
```
