## Ejercicio Extra
Programa una calculadora que

* Pregunte qué operación quieres realizar (multiplicación/división/suma/resta/potencia/salir)
* Pida dos números 
* Realice la operación
* Siga funcionando hasta que el usuario elija salir

#### Bonus:
* Que guarde todas las operaciones realizadas y las imprima por pantalla cuando el usuario elija esa opción
* Que sea robusta frente a entradas erróneas del usuario o frente a operaciones como dividir por 0

```python
import math
import pprint
import time
```

```python
# Version sin bonus y con cuidadito de meter las palabras y numeros justos

while (True):
    opcion = input ("¿Qué operación quieres realizar? Elige entre: \n Suma \n Resta \n Multiplicacion \n Division \n Potencia \n Salir").lower()
    if opcion == "salir":
        break
    operador1 = float (input ("Introduce el primer numero:"))
    operador2 = float (input ("Introduce el segundo número:"))
    match opcion:
        case "suma":
            resultado = operador1 + operador2
        case "resta":
            resultado = operador1 - operador2
        case "multiplicacion":
            resultado = operador1 * operador2
        case "division":
            resultado = operador1 / operador2
        case "potencia":
            resultado = operador1 ** operador2
    print (f"El resultado de la {opcion} es {resultado}")

print ("Programa finalizado")
    

```

```python
# Version guardando todos los resultados, metiendo las palabras justas e imprimiendo por pantalla 

log_resultados = []
while (True):
    opcion = input ("¿Qué operación quieres realizar? Elige entre: \n Suma \n Resta \n Multiplicacion \n Division \n Potencia \n Imprimir resultados \n Salir").lower()
    if opcion == "salir":
        break
    elif opcion == "imprimir":
        for i in log_resultados:
            print (i)
    else:
        operador1 = float (input ("Introduce el primer numero:"))
        operador2 = float (input ("Introduce el segundo número"))
        match opcion:
            case "suma":
                resultado = operador1 + operador2
            case "resta":
                resultado = operador1 - operador2  
            case "multiplicacion":
                resultado = operador1 * operador2  
            case "division":
                resultado = operador1 / operador2 
            case "potencia":
                resultado = operador1 ** operador2
        texto = f"El resultado de la {opcion} entre {operador1} y {operador2} es {resultado}"
        print (texto)
        log_resultados.append (texto)            

print ("Programa finalizado")
```

```python
# Version guardando todos los resultados, metiendo las palabras justas e imprimiendo por pantalla 
# Controlando errores
# Controla si divide por cero
# Controla si la opcion no es válida PERO es confuso porque antes de decirte que es erróneo te 
# pide los datos de entrada

log_resultados = []
while (True):
    opcion = input ("¿Qué operación quieres realizar? Elige entre:" \
    " \na - Suma \n b - Resta \n c- Multiplicacion \n d - Division \n e - Potencia \n f - Imprimir resultados" \
    " \n g - Salir").lower()
    
    if opcion == "salir":
        break
    elif opcion == "imprimir" or opcion == "imprimir resultados" or opcion == "imprimir los resultados":
        for i in log_resultados:
            print (i)
    else:
        operador1 = float (input ("Introduce el primer numero:").replace(",","."))
        operador2 = float (input ("Introduce el segundo número").replace(",","."))
        match opcion:
            case "suma" | "sumar":
                resultado = operador1 + operador2
            case "resta" | "restar":
                resultado = operador1 - operador2  
            case "multiplicacion" | "multiplicación" | "multiplicar":
                resultado = operador1 * operador2  
            case "division" | "dividir" | "división":
                try:
                    resultado = operador1 / operador2 
                except ZeroDivisionError:
                    print ("No se puede dividir entre Cero")
                    continue 
            case "potencia" | "potenciar":
                resultado = operador1 ** operador2
            case _:
                print ("La opcion elegida no es válida. Elige otra.")
                break
        texto = f"El resultado de la {opcion} entre {operador1} y {operador2} es {resultado}"
        print (texto)
        log_resultados.append (texto)            

print ("Programa finalizado")
```
