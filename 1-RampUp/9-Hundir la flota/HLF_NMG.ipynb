Otro detalle lógico importante

Tu lógica actual intenta:

generar un barco

comprobar si es válido

colocarlo si lo es

Pero el while está envolviendo todo el proceso, incluyendo todos los barcos:

while not correcto:
    for n in self.mis_barcos.values():

Eso significa que cada vez que vuelve al while empieza a colocar todos los barcos otra vez.

Lo normal en este tipo de algoritmos es:

intentar colocar un barco

repetir hasta que ese barco sea válido

luego pasar al siguiente barco

No repetir toda la flota desde cero.

Resumen de los problemas

1️⃣ Bucle infinito

correcto nunca cambia a True.

2️⃣ Estructura del bucle poco adecuada

el while envuelve a todos los barcos, no a cada barco.

3️⃣ Antes ya corregiste bien:

la comprobación de límites

la asignación "B".

```python
#CONSTANTES
LISTA_BARCOS = {"Portaaviones":4, "Submarino_1":3, "Submarino_2":3, "Crucero_1":2, 
                "Crucero_2":2, "Crucero_3":2, "Lancha_1":1, "Lancha_2":1, "Lancha_3":1,
                "Lancha_4":1}
TAMANIO = 10
```

```python
import time
import random
import os
```

```python
class Tablero:
    
    def __init__ (self, id:str, tamanio:int, barcos:dict):
        self.id = id
        self.size = tamanio
        self.mis_barcos = barcos
        self.tabla_barcos = []
        self.tabla_disparos = []
    
    def inicializa_tableros (self):
        #Creamos los dos vacíos
        for i in range (self.size):
            fila = []
            fila2 = []
            for j in range (self.size):
                fila.append(" ")
                fila2.append ("_")
            self.tabla_barcos.append (fila)
            self.tabla_disparos.append (fila2)
            

    def coloca_barcos_manual (self):
        self.tabla_barcos[1][0] = "B"
        self.tabla_barcos[1][1] = "B"
        self.tabla_barcos[1][2] = "B"
        self.tabla_barcos[1][3] = "B"
        self.tabla_barcos [3][3] = "B"
        self.tabla_barcos [4][3] = "B"
        self.tabla_barcos [5][3] = "B"
        self.tabla_barcos [0][5] = "B"
        self.tabla_barcos [7][4] = "B"
        self.tabla_barcos [6][4] = "B"
        self.tabla_barcos [5][4] = "B"
        self.tabla_barcos [9][9] = "B"
        self.tabla_barcos [8][9] = "B"
        self.tabla_barcos [2][5] = "B"
        self.tabla_barcos [2][6] = "B"



    def coloca_barcos_random (self):
        # A partir de una posición inicial y una orientación calcula las posiciones donde iría el barco
        # Las vamos guardando en una lista
        # Se comprueba si son válidas
        # Si es que sí, se colocan
        orientaciones = ["N","S","E","O"]
        #variable para controlar si el barco se ha podido colocar controla que esté encima de otro
        correcto = False 
        while ( not correcto):
            for n in self.mis_barcos.values():
                i = random.randint (0,TAMANIO-1)
                j = random.randint (0,TAMANIO-1)
                dir = random.choice (orientaciones)
                #Variable para guardar las posiciones
                barco = []
                barco.append ([i,j]) #guardo la posicion inicial
                if n != 1: #si el barco tiene más de una posición calculo el resto
                    for posicion in range(n-1): #ya he guardado la primera posicion, por eso quito uno. 
                        match dir:
                            case "N":
                                fila = i+1
                                col = j
                            case "S":
                                fila = i-1
                                col = j
                            case "E":
                                fila = i
                                col = j+1
                            case "O":
                                fila =i
                                col= j-1
                        barco.append ([fila,col])
                        i=fila
                        j=col
                #Compruebo que en ninguna posición de la lista hay ya un barco
                valido = True
                for i,j in barco:
                    if (0> i) or (i> 9) or (0> j) or (j > 9) or self.tabla_barcos [i][j] == "B":
                        valido = False
                        break 
                #coloca el barco
                if (valido):
                    for i, j in barco:
                        self.tabla_barcos [i][j] = "B"
            correcto = True   
    def imprime_tablero (self,selecc):
        #Uso el mismo método para imprimir los dos tableros dependiendo del parámetro
        # 1 me imprime el estado de mis barcos
        # 0 (otro) me imprime donde he disparado
        if selecc == 1:
            tablero = self.tabla_barcos
        else:
            tablero = self.tabla_disparos

        os.system('cls')
        #Printar del 0 al 9 ¿?

        for i in range (len(tablero)):
            for j in range (len (tablero[i])):
                print (f"[ {tablero[i][j]} ]", end= " ") 
            print() 


    def disparo (self, i,j):
        '''Devuelve true si el disparo ha tocado barco o False si ha tocado agua'''
        if self.tabla_barcos [i][j] == "B":
            return True
        elif self.tabla_barcos [i][j] == " ":
            return False
    
    def actualiza_tablero (self, disparo, i, j): 
        '''Qué barcos me han tocado'''
        if (disparo):
            self.tabla_barcos [i][j] = "X"
            print (f"Tocado en la posición fila: {i}, columna {j}")
        else:
            self.tabla_barcos [i][j] = "O"
            print ("Agua")

    def registra_disparo (self, disparo, i, j):
        ''' Para ver dónde voy disparando'''
        if (disparo):
            self.tabla_disparos [i][j] = "X"
        else:
            self.tabla_disparos [i][j] = "O"
    
    def es_fin_partida (self):
        for fila in self.tabla_barcos:
            if "B" in fila:
                return False
        return True 
```

```python
def pide_coordenada ():
    '''
        Pide coordenada al jugador y la devuelve 
        Comprueba que sea válida y esté en el rango de 0 a 9. Hasta que no sea así no
        finaliza la función
    '''
    while True:
        try:
            coord = int (input("Introduce la coordenada a la que quieres disparar. De 0 a 9"))
            if 0<=coord<=9:
                return coord
            else:
                print("La coordenada tiene que estar entre 0 y 9. Introdúcela de nuevo:")
        except:
            print("La coordenada tiene que estar entre 0 y 9. Introdúcela de nuevo:")

def disparo_aleatorio ():
    i = random.randint (0,9)
    j = random.randint (0,9)
    return i,j
```
