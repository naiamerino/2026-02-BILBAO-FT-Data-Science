## Mini Hundir la Flota
#### Primeros pasos
Vamos a empezar a programar trozos de código que nos servirán para el HLF:

* Crear un Tablero como una lista de listas de 10x10 que contenga caracteres espacio: " "
* Insertar dos barcos sustituyendo en el Tablero los espacios en blanco por el caracter "B":
    * Uno de 4 casillas en horizontal: casillas [1][0], [1][1], [1][2], [1][3]
    * Uno de 3 casillas en vertical:   casillas [3][3], [4][3], [5][3]
* Escribir una celda de código separada de las anteriores para simular que el usuario está efectuando un disparo: 
    *  Solicite al usuario dos coordenadas `i,j` mediante dos inputs
    *  Compruebe si esas coordenadas están ocupada por una casilla barco (que haya una en "B")
        * Si hay una "B":
            * Cambiar el contenido de esa posición por una "x"
            * Imprimir por pantalla el texto "Tocado en posición `i,j`"
        * Si hay un " ":
            * Cambiar el contenido de esa posición por una "o"
            * Imprimir por pantalla el texto "Agua"
    * Imprimir por pantalla el Tablero
* Para comprobar que funciona bien la celda de disparar ejecutarla varias veces metiendo distintas coordenadas

### Bonus
* En la celda de disparar, si el usuario acierta a tocar un barco, vuelve a disparar
* Resuélvelo con funciones:
    * Una para solicitar una coordenada válida a la que llamarás dos veces, una par i y otra par j
    * Una `disparar` que devuelva `True` si el usuario ha acertado o `False` si no.
    


```python
#Funciones definidas:

def tablero_vacio ():
    '''
    Crea y devuelve un tablero 10x10 vacío -con caracteres " "
    '''
    tablero =[]
    for i in range (10):
        fila = []
        for j in range (10):
            fila.append(" ")
        tablero.append (fila)
    return tablero

def imprimir_tablero (matriz):
    '''
        Saca por pantalla la matriz 2D que recibe como parámetro.
        Para mostrarla en modo tablero y no un elemento en cada línea sustituye el caracter
        de fin por nada " " en cada fila y salta de línea después 
    '''
    for i in range (len(matriz)):
        for j in range (len (matriz[i])):
            print (matriz[i][j], end= " ") 
        print() 


def pide_coordenada ():
    '''
        Pide y devuelve una coordenada. 
        Comprueba que sea válida y esté en el rango de 0 a 9. Hasta que no sea así no
        finaliza la función
    '''
    while True:
        try:
            coord = int (input("Introduce la coordenada a la que quieres disparar. De 0 a 9"))
            if 0<=coord<=9:
                return coord
            else:
                print("La coordenada tiene que estar entre 0 y 9. Introdúcela de nuevo:")
        except:
            print("La coordenada tiene que estar entre 0 y 9. Introdúcela de nuevo:")

def disparar (tablero, i, j):
    ''' 
        Recibe un tablero y dos coordenadas. Devuelve true si había barco y false si no
    '''
    if tablero [i][j] == "B":
        return True
    elif tablero [i][j] == " ":
        return False

    
```

```python
#Programa

tablero = tablero_vacio ()
#Insertamos barcos:
tablero[1][0] = "B"
tablero[1][1] = "B"
tablero[1][2] = "B"
tablero[1][3] = "B"
tablero [3][3] = "B"
tablero [4][3] = "B"
tablero [5][3] = "B"

turno = True
while (turno):
    fila = pide_coordenada()
    col = pide_coordenada()
    disparo = disparar (tablero, fila, col)
    if (disparo):
        tablero [fila][col] = "X"
        print (f"Tocado en posición fila:{fila} columna:{col}")
        imprimir_tablero (tablero)
    else:
        tablero [fila][col] = "O"
        print ("Agua")
        imprimir_tablero (tablero)
        turno = False 
```
