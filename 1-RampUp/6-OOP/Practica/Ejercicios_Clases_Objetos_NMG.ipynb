## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomésticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas últimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un número con todas las ventas.
* Método que calcula la media de ventas de los últimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"
2. Crear tres tiendas con datos inventados
3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)
3. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`
4. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
5. Seguro que ya has hecho este apartado... Documenta la clase :)

```python
class Tienda:
    
    '''
    Clase que representa una tienda. 
    Atributos fijos tipo = Electrodoméstico y abierta = true

    Atributos que se le pasan al constructor: 
    nombre (str)
    direccion (str)
    empleados(int) 
    ventas de los últimos tres meses (lista de tres numeros)
     
    Tiene los siguientes métodos:
    calcula_ventas(): Metodo que calcula la suma de las ventas y devuelve la suma
    ventas_por_empleado(): Metodo que calcula la media de ventas de los ultimos meses por empleado
    nombre_direccion_tienda(): Metodo que devuelve string con nombre y direccion de la tienda
    ventas_ultimo_mes(): Metodo que devuelve las ventas del ultimo mes
    proyeccion_ventas(): Metodo que devuelve la proyeccion de ventas en función de una inversión x
    
    '''
    
    tipo ="Electrodomesticos"
    Abierta =True
    
    def __init__ (self, nombre_tienda:str, dir_tienda:str, num_empleados:int, ventas_3meses:list):
        self.nombre = nombre_tienda 
        self.direccion = dir_tienda 
        self.empleados = num_empleados 
        self.ventas= ventas_3meses 

    
    def calcula_ventas (self):
        return sum (self.ventas)

    
    def ventas_por_empleado (self):
        media = self.calcula_ventas()/self.empleados
        print ("La media de ventas por empleado es:", media)

    
    def nombre_direccion_tienda (self):
        return self.nombre + " , " + self.direccion

   
    def ventas_ultimo_mes (self):
        return self.ventas [-1]

    
    def proyeccion_ventas (self, inversion):
        if inversion < 1000:
            for i, mes in enumerate(self.ventas):
                self.ventas[i] = mes*1.2
        elif inversion >= 1000:
            for i, mes in enumerate (self.ventas):
                self.ventas[i] = mes*1.5
        return self.ventas
```

## Ejercicio 2
En este ejercicio vamos a implementar una clase *Perro* en Python. La clase tiene las siguientes características:
* Cosas que sabemos seguro que tiene un perro
    * Tiene 4 patas
    * 2 orejas
    * 2 ojos
    * Una velocidad de 0. Por defecto, el perro está parado
* Cuando se inicialice:
    * El perro será de una determinada raza
    * Por defecto tendrá pelo "Marrón", a no ser que se diga lo contrario.
    * Por defecto no tendrá dueño, a no ser que se diga lo contrario.
    
* Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.
* Necesita otro método (parar), donde pondremos la velocidad a 0.
* Otro método llamado "ladrar", que tendrá un argumento de entrada, y la salida será el siguiente string: "GUAU!" + el argumento de entrada.


Se pide:
* Implementa la clase *Perro*
* Crea un objeto de tipo *Perro*, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo
* Documenta la clase *Perro*

```python
class Perro:
    '''
    Clase que recoge las características de un perro.

    Tiene los siguientes atributos:
    raza (str): Raza del perro.
    patas (int): Número de patas del perro (por defecto 4).
    orejas (int): Número de orejas del perro (por defecto 2).
    ojos (int): Número de ojos del perro (por defecto 2).
    velocidad (int | float): Velocidad actual del perro (por defecto 0).
    colorpelo (str): Color del pelo del perro (por defecto "Marron").
    tiene_amo (str | None): Nombre del amo del perro o None si no tiene.

    Tiene los siguientes métodos
    andar(): Metodo para aumentar la velocidad del perro en la cantidad indicada
    parar(): Metodo para volver a poner la velocidad del perro a cero
    ladrar(): Metodo que devuelve GUAU más el parámetro de entrada

    '''
      
    def __init__ (self, raza, num_patas = 4, num_orejas = 2, num_ojos = 2, velocidad = 0, pelo ="Marron", tiene_amo = None): 
        self.raza = raza
        self.patas = num_patas
        self.orejas = num_orejas
        self.ojos = num_ojos
        self.velocidad = velocidad
        self.colorpelo = pelo
        self.tiene_amo = tiene_amo
    

    def andar (self, aumento_velocidad):
        ''' Si ponemos aqui el comentario del método también podemos reclamarlo con Perro.andar.__doc__
        Este método aumenta el atributo velocidad en el valor pasado como parámetro'''
        self.velocidad = self.velocidad + aumento_velocidad

    
    def parar (self):
        self.velocidad = 0

    
    def ladrar (self, ladrido):
        return "GUAU!" + " " + ladrido
    
```

```python
labrador = Perro ("Labrador")
```

```python
labrador.andar (10)
print (labrador.velocidad)
```

```python
labrador.parar()
print (labrador.velocidad)
```

```python
labrador.ladrar ("Naia")
```
