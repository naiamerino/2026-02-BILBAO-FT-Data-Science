## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

$$radianes = grados*(pi/180)$$

````python
import math
````

````python
grados = float (input("Introduce los grados para convertir a radianes"))
radianes = round (grados * (math.pi/180), 2)
print (f"{grados} grados = {radianes} radianes")
````

## Ejercicio 6
Mismo programa que en 5, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
entrada1 = input ("Escribe algo")
entrada2 = input ("Escribe otra cosa")
entrada3 = input ("Escribe otra cosa más")
if (entrada1 == entrada2 == entrada3):
    print ("Todas las entradas son iguales")
elif (entrada1 == entrada2 or entrada1 == entrada3 or entrada2 == entrada3):
    print ("Al menos dos de las entradas son iguales")
else:
    print ("Todas las entradas son diferentes")
````

## Ejercicio 7
Crea un programa que recoja dos inputs. Tiene que comprobar si su suma es igual, superior o inferior a 10, cada condición por separado

````python
sumando1 = float (input("Introduce un número"))
sumando2 = float (input ("Introduce otro número"))
suma = sumando1 + sumando2
if suma > 10:
    print ("La suma es mayor que 10")
if suma < 10:
    print ("La suma es menor que 10")
if suma == 10:
    print ("La suma es igual a 10")
````

## Ejercicio 10
Para el siguiente string se pide imprimir por pantalla los siguientes casos

    "Lo primero de todo, cómo están los máquinas"

1. Pasarlo todo a mayúsculas
2. Pasarlo todo a minúsculas
3. Pasar a mayúsculas solo las iniciales de las palabras
4. Crea una lista dividiendolo por sus espacios
5. Sustituye las comas `,` por puntos y comas `;`
6. Elimina las `a` minúsculas

````python
frase = "Lo primero de todo, cómo están los máquinas"
print (frase.upper())
print (frase.lower())
print (frase.title())
frase_separada = frase.split (" ")
print (frase_separada)
print (frase.replace (",", ";"))
print (frase.replace ("a",""))
````
