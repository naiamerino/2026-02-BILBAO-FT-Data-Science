## Ejercicio 12
Encuentra los índices de los valores que no sean 0 del siguiente vector [1,2,0,0,4,0]. El output sería [1, 2 4]. [Encuentra el método apropiado en la documentación.](https://numpy.org/doc/) Encapsúlalo en una función, con un argumento de entrada de un vector.

````python
def distinto_cero (vector):
    indices = np.where (vector != 0)
    return vector [indices]

my_array = np.array ([1,2,0,0,4,0])
print (distinto_cero (my_array))
````

## Ejercicio 19
1. Crea una secuencia de números que vayan del 0 al 10.
2. Crea un nuevo array, que parta del anterior, pero sustituya por el valor -1, todos los elementos cuyo valor se encuentre entre el 3 y el 8.
3. Crea un nuevo array que niegue todos los elementos entre el 3 y el 8

````python
my_array = np.arange (0, 11)
other_array = np.where ((3 < my_array) & (my_array <8), -1, my_array )
print (other_array)
another = np.where ((3 < my_array) & (my_array <8), my_array*-1, my_array )
print (another)
````
