## Ejercicio 11
No queda muy claro cuánto de dispar, de dispersa, es la medida de IBU. Mediante estadística descriptiva (una o dos métricas, sin gráficas), razona la dispersión de la variable IBU.

```python
# Calculo la desviación estándar
dsv = beers["ibu"].std()
print (f"La varianza es {dsv}")

#Calculo el rango
rango= beers["ibu"].max() - beers["ibu"].min()
print (f"El rango es {rango}")

```

Los valores se alejan en promedio casi 26 uds de la media. Y hay una diferencia importante entre el valor más alto y el más bajo. Por lo que se entiende que la dispersión es relevante.

## Ejercicio 24
Teniendo en cuenta la media de amargor de todas las cervezas, obtén una tabla con las cervezas cuyos estados tengan una media de amargor en cerveza por encima de la media

```python
# Esto había hecho al principio. No es correcto
# beerswithbreweries [beerswithbreweries.groupby('state')['ibu'].mean()>beerswithbreweries ['ibu'].mean()]["name"]

# El groupby me da un valor para el grupo. Con eso luego no puedo ir a las filas de ese grupo, que es lo
# que quiero -sacar las cervezas dentro del  estado-. 
# Tiene que generarme las mismas filas que ese estado con ese valor para luego poder comparar.
# Eso lo hace transform

media_estados = beerswithbreweries.groupby('state')['ibu'].transform('mean')
beerswithbreweries [media_estados > beerswithbreweries['ibu'].mean()]["name"]
```
