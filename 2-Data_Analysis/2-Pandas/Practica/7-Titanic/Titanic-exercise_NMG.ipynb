24. Using [`.pivot_table()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html) to count how many women or men survived by class, call it `counted`.

Don't panic and read the documentation!




```python
#Funcionamiento de pivot_df.pivot_table(
#     index=None,      # columna(s) que quieres como filas
#     columns=None,    # columna(s) que quieres como columnas
#     values=None,     # columna sobre la que aplicar la función de agregación
#     aggfunc='mean',  # función de agregación (sum, mean, len, count...)
#     fill_value=None, # valor para reemplazar NaN
# )table
# 
counted = titanic.pivot_table (index = "pclass", columns = "gender", values="survived", aggfunc="sum" )
counted
```
