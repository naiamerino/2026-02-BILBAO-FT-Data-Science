# Extrae los dominios de los correos con una expresión regular

```python
import re

emails = '''
miguellc@thebridgeschool.tech
miguel.cruz@uc3m.es
miguel-12-2000@mi-correo.net
miguellc@s@thebridgeschool.es
'''
```

Ejemplo:
* @thebridgeschool.tech
* @uc3m.es
* @mi-correo.net
* @thebridgeschool.es

```python
#todo lo que va entre una @ y un .
# poco a poco porque si no. @con todo lo quevaya detrás
pattern = re.compile(r'@.*')

matches = pattern.finditer(emails)

for match in matches:
    print(match)
```

```python
#Creo que he tardado una hora en hacerlo

#pattern = re.compile(r'@.*[^s]') Aquí estoy buscando una @ seguido de lo que sea 
# y que no acabe en s.
pattern = re.compile('@[^s].*')
matches = pattern.finditer(emails)

for match in matches:
    print(match)
```

```python
# a-zA-Z0-9- cualquiera de estos caracteres
#+ indica uno o más
# seguido de un punto
# seguido de cualquier letra mayúscula o minúscula (1 o más)
pattern = re.compile (r'@[a-zA-Z0-9-]+\.[a-zA-Z]+')
matches = pattern.finditer(emails)

for match in matches:
    print(match)
```
