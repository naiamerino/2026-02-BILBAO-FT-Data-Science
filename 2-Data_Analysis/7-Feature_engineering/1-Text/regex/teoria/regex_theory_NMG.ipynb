# [¡VAMOS A ELLO!](https://regex101.com/)

```python
import re

text_to_search = '''abcdefghijklmnopqurtuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ
áéíóú
1234567890
Ha HaHa ?Ha
MetaCharacters (Need to be escaped):
. ^ $ * + ? { } [ ] \ | ( )
miguelcruz*com
miguelcruz.com
miguelcruzocom
miguelcruz hola
321 555 4321
123.555.1234
123*555*1234
800-555-1234
900-555-1234
9005551234
900055501234
Mr. Scha2fer
Mr Smith
Ms Davis
Mrs. Robinson
Mr.  T
Mr. ()

cat
mat
pat
bat 
at
'''
```

## Utilizamos las raw_strings para obtener la literalidad del texto:

### `print(r'\tTabulador')`

```python
print('Tabulador sin raw string: \tTabulador') #sin r, python interpreta, es un tabulador
print(r'Tabulador con raw string: \tTabulador') #con r, literal
```

### Buscamos el patrón `abc` en el texto

Para ello utilizamos:
- `re.compile()`: para introducir el patrón que queremos buscar
- La función `finditer()`: para buscar el patrón en nuestro texto
- Iteramos sobre la búsqueda

```python
#crea patrón de búsqueda
pattern = re.compile(r'abc') 

#Busca en el texto indicado las apariciones del patrón
matches = pattern.finditer(text_to_search)

for i in matches:
    print(i)


# el span es el índice de inicio y final de la coincidencia.
# gracias al span, podemos utilizar las técnicas de string slicing
# en python para localizarlo

print(text_to_search[0:3])
```

El print me da este resultado: <re.Match object; span=(0, 3), match='abc'> Me indica que hay una ocurrencia entre las posiciones 0 y 3. Me añadiría una línea como esa por cada ocurrencia

```python
#comprobación
text_to_search[0:3]
```

### Hay que tener en cuenta que cuando específicamos el pattern, se busca la literalidad de ese patrón.
Por ejemplo, si queremos buscar las letras en distinto orden...

```python
new_pattern = re.compile(r'cba')
new_matches = new_pattern.finditer(text_to_search)

for match in new_matches:
    print(match) # no se muestra nada por pantalla 
```

## Metacaracteres
Son aquellos caracteres que no son alfanuméricos:
- Signos de puntuación, exclamación y admiración

Si queremos obtenerlos, tenemos que "escaparlos"  

SIgnifican otra cosa en las búsquedas normales, p.ej . = cualquier cosa. Si lo que quiero buscar es el punto en sí tengo que 'escaparlo', que no lo interprete

```python
# Como veis, aquí se muestran prácticamente todos los caracteres, cuando lo que queremos encontrar es el punto "."
pattern = re.compile(r'.')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match) 
```

#### Para escaparlos (que no se interpreten), tienen que ir precedidos de la barra invertida(`\`)

```python
pattern = re.compile(r'\.')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

Para buscar una página web:

```python
# El punto sin interpretar me dice que ahí puede haber cualquier caracter o ninguno. Si quiero el punto tendría que poner la barra invertida
# \.
pattern = re.compile(r'miguelcruz.com')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

Lo realmente interesante de regex no es encontrar simplemente una página web o una frase concreta, sino que nos ayuda a encontrar una serie de patrones en los textos.

En este documento podemos ver las principales expresiones regulares para encontrar texto: `snippets.txt`

```python
pattern = re.compile(r'\D') #cualquier digito que no sea 0-9. Listado en el .txt de arriba

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

## Anclas

Las anclas no buscan caracteres en concreto, pero delimitan nuestra búsqueda.

Word Boundaries `\b`: está compuesto por los espacios, tabuladores, nuevas líneas y caracteres no alfanuméricos.

```python
#Quiero un Ha que esté precedido de un limitador de palabra = que sea el principio de una palabra
#Ha HaHa ?Ha 
# Encuentra tres porque el interrogante sí es un limitador de palabra
pattern = re.compile(r'\bHa')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

No word boundaries `\B`: lo contrario

Muestra el penúltimo Ha, porque delante no tiene los boundaries

```python
pattern = re.compile(r'\BHa')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

### `^` Busca solo el principio del string

```python
sentence = '''Start a sentence and end then bring it \n
Start to an end'''
```

```python
pattern = re.compile(r'^Start') #^si está fuera de corchetes significa principio de string

matches = pattern.finditer(sentence)

for match in matches:
    print(match)
    print(match[0])
```

### `$` Solo busca el final del string

```python
pattern = re.compile(r'end$')

matches = pattern.finditer(sentence)

for match in matches:
    print(match)
```

## Character sets
Sirven para concretar nuestra búsqueda.

#### ¡CUIDADO! En ocasiones suele haber confusión con los character sets, porque no cogen más de un elemento.

```python
# Para encontrar todos los números que empiecen por centenas:
# 800 - 900

pattern = re.compile(r'[89]00\D\d\d\d\D\d\d\d\d')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

## Los guiones no solamente sirven para encontrar ese caracter especial, sino que además nos permiten establecer rangos

Por ejemplo, para mostrar los números entre el 1 y el 5 de todo el texto

```python
# Aquí no estamos buscando este patrón '1-5' sino cualquier numero entre 1 y 5
# Es uno de los problemas de regex, mismos caracteres pueden ser cosas diferentes en función de qué vayan rodeados
pattern = re.compile(r'[1-5]')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

### Para Mostrar letras mayúsculas y minúsculas, basta con poner los rangos juntos.


```python
# Busca cualquier letra del alfabeto. No es exactamente igual que \D porque puede haber algún caracter que no sea alfanumérica pero que no
# esté en ese alfabeto
pattern = re.compile(r'[a-zA-Z]')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# Para él los caracteres con tilde son otro diferente. No saldría en lo anterior
pattern = re.compile(r'[á-ú]')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

## Importante 
Al poner el símbolo `^` dentro de los corchetes `[]`, significa que **NO** queremos lo que está dentro de él.

En este caso, al ejecutar, se muestran solo los caracteres numéricos, los espacios en blanco, los saltos de línea y los caracteres numéricos.

**Se niega el set**

```python
pattern = re.compile(r'[^a-zA-Z]')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

## Búsquedas de patrones en los textos 
Pongamos que queremos recoger palabras terminadas en at, excepto **bat**
Especificamos que no queremos los valores que empiecen por b

```python
pattern = re.compile(r'[^b]at')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

## Rangos `{}`
Como vemos en snippets.txt, las llaves nos permiten establecer rangos. 

Volviendo al ejemplo de los números de teléfono, otra forma de obtener los patrones

```python
#{3} - 3 apariciones
pattern = re.compile(r'\d{3}\D\d{3}\D\d{4}')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
#{2,4} entre dos y cuatro apariciones
pattern = re.compile(r'\d{2,4}.\d{2,4}.\d{2,4}')
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
## Este ejemplo nos vale porque sabemos exactamente el patrón que se reproduce.

pattern1 = re.compile(r'Mr\.')

matches = pattern1.finditer(text_to_search)

for match in matches:
    print(match)

# Aquí no nos está dando lo que queremos. Solo nos da la secuencia Mr.
```

```python
# Aquí sí aparecen todos los Mr. independientemente de que tengan punto o no
pattern2 = re.compile(r'Mr.')

matches = pattern2.finditer(text_to_search)

for match in matches:
    print(match)
```

## Operador `?` 
Nos sirve para añadir 0 o 1 a nuestra selección. Aplica al carácter que está justo a su izquierda. Así se va a contemplar lo que hay un espacio después.

```python
# Aplica sobre lo que está a su izda. Todos los Mr seguidos o no de un punto
pattern2 = re.compile(r'Mr\.?')

matches = pattern2.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
#\s Todos los Mr seguidos de . o nada seguido espacio en blanco y un \w caracter de palabra

pattern3 = re.compile(r'Mr\.?\s\w+') # El operador + muestra si hay 1 elemento o más a la derecha de la selección

matches = pattern3.finditer(text_to_search)

for match in matches:
    print(match)

# Por eso no se imprime Mr. T porque tiene dos espacios entre el. y T
```

## Ahora sí que sí
para mostrarlo todo , utilizaremos el cuantificador `*`

```python
#Rodos los Mr que tengan a continuación un punto o no, seguidos de un sólo espacio y seguido de caracteres de palabras
pattern4 = re.compile(r'Mr\.?\s\w*')

matches = pattern4.finditer(text_to_search)

for match in matches:
    print(match)
```

Los dos últimos son 
- Mr.  T así que me coge el Mr. porque con el * le estoy dando la oportunidad de que aparezcan 0 caracteres de palabra
- el otro no sé pero tb coge sólo el Mr. porque eso sí cumple y lo de detrás no

## Grouping `()`
Siguiendo con el ejemplo, para ver todos los Mr, Ms y Mrs, podemos utilizar el operador | (or)

```python
pattern4 = re.compile(r'(Mr|Ms|Mrs)\.?\s\w*')

matches = pattern4.finditer(text_to_search)

for match in matches:
    print(match)
```
